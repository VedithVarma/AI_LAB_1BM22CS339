```python
#1. TicTacToe
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board):
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]

    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]

    return None

def is_full(board):
    return all(cell != " " for row in board for cell in row)

def computer_move(board):
    empty_cells = [(r, c) for r in range(3) for c in range(3) if board[r][c] == " "]
    count=0
    x=-1
    y=-1
    for i in range(3):
      count=0
      x=-1
      y=-1
      for j in range(3):
        if board[i][j]=='O':
          count+=1
        else:
          if(board[i][j]==' '):
            x=i
            y=j
      if(count==2 and (x!=-1 and y!=-1)):
        cell=(x,y)
        if(cell in empty_cells):
          return cell
    for i in range(3):
      count=0
      x=-1
      y=-1
      for j in range(3):
        if board[j][i]=='O':
          count+=1
        else:
          if(board[j][i]==' '):
            x=j
            y=i
      if(count==2 and (x!=-1 and y!=-1)):
         cell=(x,y)
         if(cell in empty_cells):
           return cell
    count=0
    x=-1
    y=-1
    for i in range(3):
      if board[i][i]=='O':
        count+=1
      else:
        if(board[i][i]==' '):
          x=i
          y=i
    if(count==2 and (x!=-1 and y!=-1)):
      cell=(x,y)
      if(cell in empty_cells):
        return cell
    count=0
    x=-1
    y=-1
    for i in range(3):
      if board[i][3-1-i]=='O':
        count+=1
      else:
        if(board[i][3-1-i]==' '):
          x=i
          y=i
    if(count==2 and (x!=-1 and y!=-1)):
      cell=(x,y)
      if(cell in empty_cells):
        return cell

    for i in range(3):
      count=0
      x=-1
      y=-1
      for j in range(3):
        if board[i][j]=='X':
          count+=1
        else:
          if(board[i][j]==' '):
            x=i
            y=j
        if(count==2 and (x!=-1 and y!=-1)):
          cell=(x,y)
          if(cell in empty_cells):
            return cell
    for i in range(3):
      count=0
      x=-1
      y=-1
      for j in range(3):
        if board[j][i]=='X':
          count+=1
        else:
          if(board[j][i]==' '):
            x=j
            y=i
        if(count==2 and (x!=-1 and y!=-1)):
          cell=(x,y)
          if(cell in empty_cells):
            return cell
    count=0
    x=-1
    y=-1
    for i in range(3):
      if board[i][i]=='X':
        count+=1
      else:
        if(board[i][i]==' '):
          x=i
          y=i
    if(count==2 and (x!=-1 and y!=-1)):
      cell=(x,y)
      if(cell in empty_cells):
        return cell
    count=0
    x=-1
    y=-1
    for i in range(3):
      if board[i][3-1-i]=='X':
        count+=1
      else:
        if(board[i][3-1-i]==' '):
          x=i
          y=i
    if(count==2 and (x!=-1 and y!=-1)):
      cell=(x,y)
      if(cell in empty_cells):
        return cell
    return random.choice(empty_cells)

def tic_tac_toe():
    board = []
    for _ in range(3):
        row = []
        for _ in range(3):
            row.append(" ")
        board.append(row)

    current_player = "X"  # Player X is the human

    while True:
        print_board(board)

        if current_player == "X":
            row = int(input("Player X, enter your row (0-2): "))
            col = int(input("Player X, enter your column (0-2): "))
        else:
            row, col = computer_move(board)
            print("Computer (O) chooses row " + str(row) + ", column " + str(col))

        if board[row][col] == " ":
            board[row][col] = current_player
        else:
            print("Cell already taken! Try again.")
            continue

        winner = check_winner(board)
        if winner:
            print_board(board)
            print("Player " + winner + " wins!")
            break

        if is_full(board):
            print_board(board)
            print("It's a draw!")
            break

        current_player = "O" if current_player == "X" else "X"

tic_tac_toe()
```

```python
#2. Vacuum Cleaner
import random
environment = {
    "A": random.choice(["Dirty", "Clean"]),
    "B": random.choice(["Dirty", "Clean"])
}

vacuum_location = random.choice(["A", "B"])

def display_environment():
    print("Location A is", environment['A'])
    print("Location B is", environment['B'])
    print("Vacuum is at location", vacuum_location)
    print()

def vacuum_agent():
    global vacuum_location

    while environment["A"] == "Dirty" or environment["B"] == "Dirty":
        display_environment()
        if environment[vacuum_location] == "Dirty":
            print("Cleaning location", vacuum_location)
            print()
            environment[vacuum_location] = "Clean"
        else:
            print("Location", vacuum_location, "is clean. Moving to the other location...")
            print()
            vacuum_location = "A" if vacuum_location == "B" else "B"

    print("Both locations are clean. Task completed!")

vacuum_agent()
```
